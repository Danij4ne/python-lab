## Loop reading and line cleaning

```python
with open("text.txt", "r") as my_file:
    while True:
        line = my_file.readline()
        if not line:
            break
        print(line.strip())
```

```python
with open("text.txt", "r") as my_file:
    i = 0
    for line in my_file : 
        print("my interactions", i, ":" , line.strip())
        i+=1
```

```python
with open("text.txt", "r") as my_file:
    i = 0
    for line in my_file : 
        if line.strip() :
            print("my interactions", i, ":" , line.strip())
            i+=1
```

```python
with open("text.txt", "r") as my_file:
    counter = 0
    for word in my_file : 
        my_words = len(word.split())
        counter+=my_words
    print("total words : " , counter) 
```

```python
with open("text.txt","a") as my_file :
    my_file.write("important")
with open("text.txt","r") as my_file :
    for line in my_file:
        if "important" in line.lower():
            print(line.strip())
        
```

## Copy file line by line

```python
with open ("lista.txt","r") as my_list , open ("copia.txt", "w") as my_copia:
    for line in my_list:
        my_copia.write(line)
```

## Merge files with separator

```python
with open ("lista.txt","r") as my_list , open ("copia.txt", "r") as my_copia , open("fusion.txt","w") as copia:
    for line in my_list:
        copia.write(line)
        
    copia.write("\n--- SEPARADOR ---\n")
    
    for line in my_copia:
        copia.write(line)

with open("fusion.txt", "r") as fusion_result:
    print(fusion_result.read())

    
```

## Temporary copy and replacement

```python
import shutil
import os

# Copiar archivo
shutil.copy("fusion.txt", "copyfusion.txt")

# Verificar si existe la copia
if os.path.exists("copyfusion.txt"):
    print("✅ El archivo se copió correctamente.")
else:
    print("❌ No se encontró el archivo de copia.")

```

## Merge avoiding duplicates


```python
file1 = "copyfusion.txt"
file2 = "fusion.txt"
file3 = "newfusion.txt"

with open (file1 , "r", encoding="UTF-8") as f1:
    newf1 = set(f1.readlines())

with open (file2 , "r", encoding="UTF-8") as f2:
    newf2 = (f2.readlines())


    for line in newf2 :
        if line not in newf1:
            newf1.add(line)

with open (file3 , "w" , encoding="UTF-8") as f3:
    f3.writelines(sorted(newf1)) 
```
