## XML

```python
import xml.etree.ElementTree as etree
```

```python
tree = etree.parse("employees.xml")
root = tree.getroot()
```

```python
rows = [] 
```

```python
for node in root:
    name = node.find("name").text
    role = node.find("role").text
    salary = node.find("salary").text
    rows.append([name, role, salary])
```

```python
df = pd.DataFrame(rows, columns=["name", "role", "salary"])
df
```
