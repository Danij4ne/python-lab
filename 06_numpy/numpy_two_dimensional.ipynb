# Two Dimensional Numpy

```python
import numpy as np
```

```python
a = [[11,12,13], [21,22,23], [31,32,33]]
A = np.array(a) #creamos un array de numpy , ahora si es 2D
A
```

```python
A.ndim # 2D
```

```python
A.shape #forma array 3 filas , 3 columnas
```

```python
A.size #tamaño
```

```python
A[1][2] 
```

```python
x = np.array([[1,0],[1,0]])
y = np.array([[2,1],[3,0]])
r = x + y
r
```

```python
rm = x * y
rm
```

```python
m = np.array([[4,3,9],[5,1,3]])
n = np.array([[4,3],[5,1],[7,2]])
mn = np.dot(m,n)
mn
```

```python
arr = np.array([4, 9, 16, 25])
rz = result = np.sqrt(arr) # calcula la raíz cuadrada de cada elemento del array
rz
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)  #  cambia la forma del array a 2 filas × 3 columnas
reshaped
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
transposed_arr = arr.T    # convierte filas en columnas y columnas en filas
print(transposed_arr)
```
