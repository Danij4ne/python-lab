## Vectores con matplotlib

```python
u = np.array([1, 0])
v = np.array([0, 1])
z = u + v
def Plotvec1(a, b, c):
    origin = np.zeros(2)  # (0,0)
    plt.quiver(*origin, *a, angles='xy', scale_units='xy', scale=1, color='r', label='u')
    plt.quiver(*origin, *b, angles='xy', scale_units='xy', scale=1, color='g', label='z')
    plt.quiver(*origin, *c, angles='xy', scale_units='xy', scale=1, color='b', label='v')

    plt.xlim(-1, 3)
    plt.ylim(-1, 3)
    plt.grid(True)
    plt.legend()
    plt.show()


u = np.array([1, 0])
v = np.array([0, 1])
z = u + v


Plotvec1(u, z, v)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear 100 valores entre 0 y 10
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Primer grfico (izquierda)
plt.subplot(1, 2, 1)   # 1 fila, 2 columnas, primer grfico
plt.plot(x, y_sin)

# Segundo grfico (derecha)
plt.subplot(1, 2, 2)   # 1 fila, 2 columnas, segundo grfico
plt.plot(x, y_cos)

plt.show()
```
