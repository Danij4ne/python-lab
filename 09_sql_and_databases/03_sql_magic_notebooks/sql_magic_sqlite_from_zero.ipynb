# SQL Magic + SQLite from Scratch in Jupyter


## Enable SQL Magic in Jupyter

```python
%load_ext sql
```

## Connect to (and create if it does not exist) a SQLite database

```python
%sql sqlite:///archivo.db
```

## Create a table

```python
%%sql
CREATE TABLE IF NOT EXISTS instructors (
    id INTEGER PRIMARY KEY,
    firstname TEXT,
    lastname TEXT,
    city TEXT,
    country TEXT
);
```

## Insert data

```python
%%sql
INSERT INTO instructors (id, firstname, lastname, city, country)
VALUES
(1, 'John', 'Doe', 'New York', 'USA'),
(2, 'Jane', 'Smith', 'London', 'UK'),
(3, 'Carlos', 'Gomez', 'Madrid', 'Spain');
```

## Query data

```python
%%sql
SELECT * FROM instructors;

```

```python
%%sql
SELECT firstname, lastname
FROM instructors
WHERE country = 'Spain';
```

## Store results in a Python variable

```python
%%sql result <<
SELECT * FROM instructors;
```
