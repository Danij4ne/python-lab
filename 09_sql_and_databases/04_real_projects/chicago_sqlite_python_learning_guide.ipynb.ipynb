## import CSV files and store them in variables using pandas


```python
import pandas as pd 
census_data = pd.read_csv("chicagocensusdata.csv")
crime_data = pd.read_csv("chicagocrimedata.csv") 
school_data = pd.read_csv("chicagopublicschools.csv") 
```

## Save the CSV file as SQL

```python
census_data.to_sql(
    name = "census_data" , 
    con = "sqlite:///chicago.db",
    if_exists = "replace",
    index = False 
)
```

```python
crime_data.to_sql(
    name = "crime_data" ,
    con = "sqlite:///chicago.db" ,
    if_exists = "replace",
    index = False
)
```

```python
school_data.to_sql(
    name = "school_data", 
    con = "sqlite:///chicago.db" ,
    if_exists = "replace" ,
    index = False
)
```

```python
%%sql 
SELECT *
FROM census_data
```

```python
%%sql 
SELECT COMMUNITY_AREA_NAME , PERCENT_AGED_16__UNEMPLOYED 
FROM census_data
WHERE PERCENT_AGED_16__UNEMPLOYED > 8
ORDER BY  PERCENT_AGED_16__UNEMPLOYED DESC
```

## Get the communities whose poverty index is above the average

```python
%%sql 
SELECT COMMUNITY_AREA_NAME , AVG(HARDSHIP_INDEX) AS avg_hardship
FROM census_data
GROUP BY COMMUNITY_AREA_NAME
HAVING AVG(HARDSHIP_INDEX) > (SELECT AVG(HARDSHIP_INDEX) FROM census_data )
ORDER BY avg_hardship DESC
LIMIT 10
```

## 5 least frequent crime types

```python
%%sql
SELECT
  PRIMARY_TYPE,
  COUNT(*) AS crime_count
FROM crime_data
GROUP BY PRIMARY_TYPE
ORDER BY crime_count ASC
LIMIT 5; 
```

## How many crimes have been recorded per community

```python
%%sql
SELECT COMMUNITY_AREA_NUMBER AS COMMUNITY , COUNT(*) AS crime
FROM crime_data
GROUP BY COMMUNITY_AREA_NUMBER
```
