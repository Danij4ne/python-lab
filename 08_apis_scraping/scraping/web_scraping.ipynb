```python
import requests                     
from bs4 import BeautifulSoup
```

```python
url = "https://quotes.toscrape.com/"
```

```python
resp = requests.get(url)
```

```python
print("Status code:", resp.status_code)
```

```python
html = resp.text
html
```

```python
print(html[:500])
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
type(soup)
```

```python
quote_spans = soup.find_all("span", class_="text")
print("Número de frases encontradas:", len(quote_spans))
```

```python
for i, span in enumerate(quote_spans, start=1):
    frase = span.get_text(strip=True)
    print(f"{i}. {frase}")
```

```python
primer_autor = soup.find("small", class_="author")
nombre_autor = primer_autor.get_text(strip=True)
print(nombre_autor)
```

```python
import requests
import pandas as pd

resp = requests.get("https://dummyjson.com/users?limit=5")
data = resp.json()
```

```python
usuarios = data["users"]
usuarios
```

```python
lista_limpia = []

for u in usuarios:
    lista_limpia.append({
        "firstName": u["firstName"],
        "age": u["age"],
        "email": u["email"]
    })
```

```python
df = pd.DataFrame(lista_limpia)
print(df)
```

```python
pw = requests.get("https://books.toscrape.com/")
pw.text
```

```python
soup = BeautifulSoup(pw.text, "html.parser")
link = soup.find_all("a", href= True)
link
```

```python
titles = soup.find_all("title")
titles
```

```python
titles_clean = [t.get_text().strip() for t in titles]
```

```python
titles_df = pd.DataFrame(titles_clean , columns=["title"])
titles_df
```

```python
mts = soup.find_all("meta")
mts
```

```python
mts_clean = [(t.get("content") , t.get("name") )  for t in mts]
mts_clean
```

```python
mts_df = pd.DataFrame(mts_clean, columns =["name", "content"])
mts_df
```
