### Boolean filter (True/False condition)

```python
df1 = pd_songs["Released"] > 1979
print(df1)
```

### Filter DataFrame with condition



```python
df1 = pd_songs[pd_songs["Released"] > 1979]
print(df1)
```

### Unique values Released



```python
pdr = pd_songs["Released"].unique()
print(pdr)
```

### Save the new DataFrame df1 to a file



```python
df1.to_csv("new_songs.csv", index=False)
new_songs = pd.read_csv("new_songs.csv")
new_songs
```
